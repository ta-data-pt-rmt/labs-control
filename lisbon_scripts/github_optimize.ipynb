# Week 1

```python
labs_week1 = {'lab-list-comprehensions' : 'B3',
            'lab-tuple-set-dict' : 'B4',
            'lab-string-operations' : 'B5',
            'lab-numpy' : 'B6',
            'lab-intro-pandas' : 'B7'}
```

```python
for labs in labs_week1:
    print(labs)
    time.sleep(5)
    response = requests.get('https://api.github.com/repos/ta-data-lis/' + labs + '/pulls')
    data = pd.DataFrame(response.json())
    if 'title' in data:
        pr = data['title']
        pr_name = [a.split(']')[1].strip() for a in pr if len(a.split(']')) > 1]

        # print(pr)

        names = []

        for x in alias:
            for a in pr_name:
                if a.lower().find(x) != -1:
                    names.append(alias[x])

        print(names)

        book = openpyxl.load_workbook('labs.xlsx')

        for name in names:
            sheet = book[name]
            sheet[labs_week1[labs]] = 'Delivered'

        book.save('Labs.xlsx')
```
